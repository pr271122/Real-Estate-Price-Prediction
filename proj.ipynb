```python
def remove_pps_outliers(df):
    final_df=pd.DataFrame()
    for key,new_df in df.groupby('location'):
        mean=new_df.price_per_sq_ft.mean()
        std=new_df.price_per_sq_ft.std()
        new_dff=new_df[(new_df.price_per_sq_ft>=(mean-std)) & (new_df.price_per_sq_ft<=(mean+std))]
        final_df=pd.concat([final_df,new_dff],ignore_index=True)
    return final_df
```

```python
df=remove_pps_outliers(df)
```

```python
import matplotlib
def scatter_chart(df,location):
    a=df[(df.location==location) & (df.bedroom==2)]
    b=df[(df.location==location) & (df.bedroom==3)]
    matplotlib.rcParams['figure.figsize']=(15,10)
    plt.scatter(a.total_sqft,a.price,color='blue',label='2 BHK')

    plt.scatter(b.total_sqft,b.price,color='red',label='3 BHK',marker='+')
    plt.xlabel("Total sq.ft area")
    plt.ylabel('Price')
    plt.legend()
```

```python
scatter_chart(df,'Rajaji Nagar')
```

```python
def remove_bedroom_outliers(df):
    remove_indices=np.array([])
    for location,new_df in df.groupby('location'):
        bhk_stats={}
        for bhk,new_dff in new_df.groupby('bedroom'):
            bhk_stats[bhk]={
                'mean':new_dff.price_per_sq_ft.mean(),
                'std':new_dff.price_per_sq_ft.std(),
                'count':new_dff.shape[0]
            }
        for bhk,new_dff in new_df.groupby('bedroom'):
            stats=bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                remove_indices=np.append(remove_indices,new_dff[new_dff.price_per_sq_ft<(stats['mean'])].index.values)
    return df.drop(remove_indices,axis='index')

```

```python
from sklearn.tree import DecisionTreeRegressor
diff_models={
    'Lasso':{
        'model':Lasso(),
        'params':{
            'alpha':[1,2],
            'selection':['random','cyclic']
        }
    },
    'linear_regression':{
        'model':LinearRegression(),
        'params' : {

        }
    },
    'tree':{
        'model':DecisionTreeRegressor(),
        'params':{
            'criterion':['friedman_mse'],
            'splitter':['best','random']
        }
    }
}
```

```python
from sklearn.model_selection import GridSearchCV
scores = []
cv=ShuffleSplit(n_splits=5,test_size=0.2,random_state=0)
for model_name, mp in diff_models.items():
    clf =  GridSearchCV(mp['model'], mp['params'], cv=cv, return_train_score=False)
    clf.fit(X,Y)
    scores.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })
new_df = pd.DataFrame(scores,columns=['model','best_score','best_params'])
new_df
```

```python
def predict_price(model,location,sq_ft,bath,balcony,bedroom,area):
    try:
        loc_index=np.where(X.columns==location)[0][0]
    except:
        loc_index=-1
    try:
        loc_index2=np.where(X.columns==area)[0][0]

    except:
        loc_index2=-1
    x=np.zeros(len(X.columns))
    x[0]=sq_ft
    x[1]=bath
    x[2]=balcony
    x[3]=bedroom
    if loc_index>=0:
        x[loc_index]=1
    if loc_index2>=0:
        x[loc_index2]=1
    return model.predict([x])[0]
```

```python
predict_price(model,'1st Block Jayanagar',1875,2,3,3,'Super built-up  Area')
```

```python
predict_price(model,'Indira Nagar',1000, 3, 3,3,'Super built-up  Area')
```
